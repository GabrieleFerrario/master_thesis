```python
import json
 
f = open('dataset_flickr30k.json')

flicker_karpathy = json.load(f)
```

```python
flicker_karpathy.keys()
```

```python
karpathy_split = {"train": [], "val": [], "restval": [], "test": []}
for elem in flicker_karpathy["images"]:
    karpathy_split[elem["split"]].append(elem)
```

```python
len(karpathy_split["train"])
```

```python
result = {
    "annotations" : [],
    "images": [],
    "type": "captions",
    "info": "dummy",
    "licenses": "dummy"
}
```

```python
for img in tqdm(karpathy_split["train"]):
    image_id = str(img["filename"].replace(".jpg", ""))
    for caption in img["sentences"]:
        result["annotations"].append({"image_id": image_id, "id": caption["sentid"], "caption": caption["raw"]})
    result["images"].append({"id": image_id, "file_name": image_id})
```

```python
result["annotations"]
```

```python
len(result["images"])
```

```python
f = open('train_caption_coco_format.json', 'w')

json.dump(result, f)
```

```python
f = open('train_caption.json', 'w')

json.dump(result["annotations"], f)
```
